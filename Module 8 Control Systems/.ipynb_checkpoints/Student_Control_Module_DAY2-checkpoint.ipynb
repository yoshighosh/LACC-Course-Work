```python
# Design a way-point navigation controller for a vehicle
```

```python
beta = 15 # daming coefficient
mass = 125 # Kg
x0 = 25 # meters
v0 = 0 # meters/second
force0 = 0 # Newtons, initial engine force on vehicle

A = matrix([0, 1, 0, -beta/mass]).reshape((2,2))
B = matrix([0, 1/mass]).reshape((2,1))
Q = matrix([10, 0, 0, 0]).reshape((2,2))
R = matrix([0.1])

P = CARE(A, B, Q, R)
K = R.I*B.T*P
An = A - B*K

T = np.arange(0, 30, .1)
U = [9.81 for i in T]
X0 = [x0, v0]

ss1 = sig.StateSpace(An, B, C, D)
sim = sig.lsim(ss1, U, T, X0)
pos = sim[2]
plt.plot(pos) #blue is position, orange is velocity
```

```python
# Design a Cruise Controller (Velocity tracker) for the Dynamic Car Model
```

```python
# 1) Draw FBD, 2) Write State Space, 3) Put into SS object, 
## 4) Run CARE to find optimal control law, 5) Plot Position and Velocity (see if controller works)
```

```python
beta = 15 # daming coefficient
mass = 125 # Kg
v0 = 20 # meters/second
v_goal = 30 # meters/second
force0 = 0 # Newtons, initial engine force on vehicle

A = matrix([-beta/mass, 0, 0, 1/v_goal]).reshape((2, 2))
B = matrix([1/mass, 0]).reshape((2,1))
Q = matrix([100, 0, 0, 0]).reshape((2,2))
R = matrix([.01])
Ae = matrix(A * (v0 - v_goal))
Be = matrix(B * (v0 - v_goal))
Pe = matrix(CARE(Ae, Be, Q, R))
Ke = R.I*Be.T*Pe

Ke = matrix(Ke).reshape(2,1)
K = (Ke[0] + Ke[1]*v0) - (Ke[0] + Ke[1]*v_goal)

An = A - B*K

T = np.arange(0, 20, .1)
U = [force0 for i in T]
X0 = [v0, v_goal]

ss1 = sig.StateSpace(An, B, C, D)
sim = sig.lsim(ss1, U, T, X0)
pos = sim[2]
plt.plot(pos) #blue is position, orange is velocity
```
